```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#Calculate total number of schools
tot_schools = school_data_complete["school_name"].nunique()
tot_schools
```

```python
#Calculate total students
tot_students = school_data_complete["Student ID"].nunique()
tot_students
```

```python
#Calculate total budget
tot_budget = school_data["budget"].sum()
tot_budget
```

```python
#Calculate average math score
avg_mathscore = student_data["math_score"].mean()
avg_mathscore
```

```python
#Calculate average reading score
avg_readingscore = student_data["reading_score"].mean()
avg_readingscore
```

```python
#Calculate percent of students with passing math score
passing_math_df = student_data.loc[student_data["math_score"]>=70]

passing_math_pct = passing_math_df["student_name"].count() / student_data["student_name"].count()
passing_math_pct
```

```python
#Calculate percent of students with passing reading score
passing_reading_df = student_data.loc[student_data["reading_score"]>=70]

passing_reading_pct = passing_reading_df["student_name"].count() / student_data["student_name"].count()
passing_reading_pct
```

```python
#Calculate percent of students passing both reading and math
overall_passing_df = student_data.loc[(student_data["math_score"]>=70) & (student_data["reading_score"]>=70)]

overall_passing_pct = overall_passing_df["student_name"].count() / student_data["student_name"].count()
overall_passing_pct
```

```python
#Create district summary table
district_summary_df = pd.DataFrame([{"Total Schools": tot_schools, 
                                     "Total Students": tot_students, 
                                     "Total Budget": tot_budget, 
                                     "Average Math Score": avg_mathscore, 
                                     "Average Reading Score": avg_readingscore, 
                                     "Percent of Students Passing Math": passing_math_pct*100, 
                                     "Percent of Students Passing Reading": passing_reading_pct*100, 
                                     "Percent of Students Passing Math and Reading": overall_passing_pct*100}])

district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#Calculate Total Students per school

students_per_school = school_data_complete["school_name"].value_counts()
students_per_school
```

```python
#calculate percentage of students passing math in each school
only_passing_math = school_data_complete.loc[school_data_complete["math_score"]>=70]

grouped_passing_math_df = only_passing_math.groupby("school_name")
grouped_passing_math_df.count()

passing_math_school = grouped_passing_math_df["math_score"].count()
passing_math_school

pct_passing_math_school = (passing_math_school / students_per_school)*100
pct_passing_math_school
```

```python
#Calculate percent of students passing reading per school
only_passing_reading = school_data_complete.loc[school_data_complete["reading_score"]>=70]

grouped_passing_reading = only_passing_reading.groupby("school_name")
grouped_passing_reading.count()

passing_reading_school = grouped_passing_reading["reading_score"].count()
passing_reading_school

pct_passing_reading_school = (passing_reading_school / students_per_school)*100
pct_passing_reading_school
```

```python
#Calculate percent of students passing both math and reading per school
only_overall_passing_df = school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70)]


grouped_overall_passing_school = only_overall_passing_df.groupby("school_name")
grouped_overall_passing_school.count()

overall_passing_school = grouped_overall_passing_school["student_name"].count()

pct_overall_passing_school = (overall_passing_school / students_per_school)*100
pct_overall_passing_school
```

```python
grouped_school_df = school_data_complete.groupby(["school_name"])

#calculate desired values for each school using the groupby
school_size = grouped_school_df["Student ID"].count()
school_size

school_budget = grouped_school_df["budget"].mean()
school_budget_reformat = school_budget.map("${:,.2f}".format)

budget_per_student = grouped_school_df["budget"].mean()/grouped_school_df["Student ID"].count()
budget_per_student_reformat = budget_per_student.map("${:,.2f}".format)

school_avg_mathscore = grouped_school_df["math_score"].mean()
school_avg_mathscore

school_avg_readingscore = grouped_school_df["reading_score"].mean()
school_avg_readingscore

school_type = grouped_school_df["type"].max()
school_type
```

```python
#first, save a version of the data frame without formatting the "total budget" or "per student budget" columns as strings (for use in subsequent questions)
school_summary_df = pd.DataFrame({"School Type": school_type,
                                  "Total Students": students_per_school,
                                  "Total School Budget": school_budget,
                                  "Per Student Budget": budget_per_student,
                                  "Average Math Score": school_avg_mathscore,
                                  "Average Reading Score": school_avg_readingscore,
                                  "% Passing Math": pct_passing_math_school,
                                  "% Passing Reading": pct_passing_reading_school,
                                  "% Overall Passing": pct_overall_passing_school})

#create reformatted version to improve appearance of currency values
school_summary_df_reformat = pd.DataFrame({"School Type": school_type,
                                  "Total Students": students_per_school,
                                  "Total School Budget": school_budget_reformat,
                                  "Per Student Budget": budget_per_student_reformat,
                                  "Average Math Score": school_avg_mathscore,
                                  "Average Reading Score": school_avg_readingscore,
                                  "% Passing Math": pct_passing_math_school,
                                  "% Passing Reading": pct_passing_reading_school,
                                  "% Overall Passing": pct_overall_passing_school})

school_summary_df_reformat
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# Calculate avg math score for each grade in each school

#9th graders
ninth_grade_df = school_data_complete.loc[school_data_complete["grade"]=="9th"]
ninth_grade_grouped_school = ninth_grade_df.groupby("school_name")

avg_ninth_grade_mathscore = ninth_grade_grouped_school["math_score"].mean()
avg_ninth_grade_mathscore

#10th grade
tenth_grade_df = school_data_complete.loc[school_data_complete["grade"]=="10th"]
tenth_grade_grouped_school = tenth_grade_df.groupby("school_name")

avg_tenth_grade_mathscore = tenth_grade_grouped_school["math_score"].mean()
avg_tenth_grade_mathscore

#11th grade
elev_grade_df = school_data_complete.loc[school_data_complete["grade"]=="11th"]
elev_grade_grouped_school = elev_grade_df.groupby("school_name")

avg_elev_grade_mathscore = elev_grade_grouped_school["math_score"].mean()
avg_elev_grade_mathscore

#12th grade
twel_grade_df = school_data_complete.loc[school_data_complete["grade"]=="12th"]
twel_grade_grouped_school = twel_grade_df.groupby("school_name")

avg_twel_grade_mathscore = twel_grade_grouped_school["math_score"].mean()
avg_twel_grade_mathscore

#Create new summary data frame for average math scores in each grade for each school

school_grade_mathscore_sum_df = pd.DataFrame({"9th": avg_ninth_grade_mathscore,
                                              "10th": avg_tenth_grade_mathscore,
                                              "11th": avg_elev_grade_mathscore,
                                              "12th": avg_twel_grade_mathscore})

school_grade_mathscore_sum_df
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#Create bins and group names, use "cut" to execute binning
bins = [0,584.999, 629.999, 644.999, 680]
group_names = ["< $585", " $585 - 630", " $630 - 645", " $645 - 680"]

school_summary_df["Spending Ranges (Per Student)"] = pd.cut(school_summary_df["Per Student Budget"], bins, labels=group_names, include_lowest=True)
```

```python
#Create new GroupBy to identify averages by Per Student Budget levels
schools_grouped_spending = school_summary_df.groupby("Spending Ranges (Per Student)")

#calculate average math scores
avg_mathscore_spending = schools_grouped_spending["Average Math Score"].mean()

#calculate average reading scores
avg_readingscore_spending = schools_grouped_spending["Average Reading Score"].mean()
avg_readingscore_spending

#calculate average % passing math
avg_pct_passing_math_spending = schools_grouped_spending["% Passing Math"].mean()
avg_pct_passing_math_spending

#calculate average % passing reading
avg_pct_passing_reading_spending = schools_grouped_spending["% Passing Reading"].mean()
avg_pct_passing_reading_spending

#calculate average % passing math AND reading

avg_pct_overall_passing_spending = schools_grouped_spending["% Overall Passing"].mean()
avg_pct_overall_passing_spending

#Create new summary data frame

scores_summary_spending_df = pd.DataFrame({"Average Math Score": avg_mathscore_spending, 
                                          "Average Reading Score": avg_readingscore_spending, 
                                          "% Passing Math": avg_pct_passing_math_spending, 
                                          "% Passing Reading": avg_pct_passing_reading_spending, 
                                          "% Overall Passing": avg_pct_overall_passing_spending})
scores_summary_spending_df
```

## Scores by School Size

* Perform the same operations as above, based on school size.

```python
#Create bins and group labels, use "cut" to execute binning
bins2 = [0, 999, 1999, 5000]
group_labels2 = ["Small (<1000)", "Medium (1000-2000)", "Large (2000 - 5000)"]

school_summary_df["School Size"] = pd.cut(school_summary_df["Total Students"], bins2, labels=group_labels2, include_lowest=True)
```

```python
#Create new GroupBy to find desired values for each size of school

schools_grouped_size = school_summary_df.groupby("School Size")

#Calculate average math score by school size
avg_math_score_size = schools_grouped_size["Average Math Score"].mean()

#Average reading score by schools size
avg_reading_score_size = schools_grouped_size["Average Reading Score"].mean()

#Average % Passing Math
avg_pct_passing_math_size = schools_grouped_size["% Passing Math"].mean()

#Average % Passing Reading
avg_pct_passing_reading_size = schools_grouped_size["% Passing Reading"].mean()
avg_pct_passing_reading_size

#Average % Overall Passing

avg_pct_overall_passing_size = schools_grouped_size["% Overall Passing"].mean()
avg_pct_overall_passing_size

#Create new dataframe to summarize scores and passing rates by school size
scores_summary_size_df = pd.DataFrame({"Average Math Score": avg_math_score_size, 
                                      "Average Reading Score": avg_reading_score_size, 
                                      "% Passing Math": avg_pct_passing_math_size, 
                                      "% Passing Reading": avg_pct_passing_reading_size, 
                                      "% Overall Passing": avg_pct_overall_passing_size})

scores_summary_size_df
```
